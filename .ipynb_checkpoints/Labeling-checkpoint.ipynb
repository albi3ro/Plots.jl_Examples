# Backend Warning

`gr` did not support `formatter`, so I'm switching to pyplot at this point.  

```julia
pyplot()
plot(x,y1,formatter=:scientific)
```

```julia
function mylabel(num::Real)
    if isapprox(num,Ï€)
        return "Ï€"
    elseif isapprox(num,0)
        return "0"
    end
    
    num2=num/Ï€
    
    for ii in 1:10
        if isapprox(num2,ii)
            return "$ii Ï€"
        elseif isapprox(num,-ii)
            return "-$ii Ï€"
        end
    end
        
    return "$num"
end
```

```julia
pyplot()
plot(x,y1,
    xticks=[0,Ï€,2Ï€,3Ï€,4Ï€],
    formatter=mylabel)
```
